```python
import pandas as pd
import json
```

```python
with open("parsed.json", "r") as f:
    parsed = json.load(f)["parsed"]
```

```python
sum([len(day.values()) for day in parsed])
```

```python
def create_multi_index_column(field_title):
    """Crea multi index desarmando el título de un campo."""
    
    tc, ars, coin, entity, channel, flow, hour = field_title.split("_")

    return (coin, entity, channel, flow, hour)
```

```python
df = pd.DataFrame(parsed).set_index("indice_tiempo")
df.head()
```

```python
len(df)
```

```python
df.columns = pd.MultiIndex.from_tuples([create_multi_index_column(col) for col in df.columns])
df.head()
```

```python
len(df), len(df.columns), len(df)*len(df.columns)
```

```python
df_panel = df.stack([-5, -4, -3, -2, -1], dropna=False).reset_index()
df_panel.columns = ["indice_tiempo", "coin", "entity", "channel", "flow", "hour", "value"]
df_panel.head()
```

```python
len(df_panel)
```

```python
# usar nombres de columnas en español
df_panel.columns = ["indice_tiempo", "moneda", "entidad_bancaria", "canal", "flujo", "hora", "valor"]
```

```python
# usar estas lambdas para formatear correctamente el tiempo y el valor
df_panel["indice_tiempo"] = df_panel["indice_tiempo"].apply(lambda x: x.format("Y%-M%-d%")) # esto hay que revisarlo bien!!!
df_panel["valor"] = df_panel["valor"].apply(lambda x: x if x else pd.np.nan)
```

```python
df_panel.to_csv("tipo-cambio-minorista-cotizaciones-panel.csv", encoding="utf-8", index=False, sep=",")
```
