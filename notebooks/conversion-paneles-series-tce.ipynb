# Ejemplo de conversión de paneles a series usando pandas

```python
import pandas as pd
```

```python
# leo el panel
df_panel = pd.read_csv("../datos/tce/tipos-cambio-minorista-cotizaciones-panel.csv")
df_panel.head()
```

```python
def create_field_title(col_multi_index, coin):
    """Convierte columnas muli index a nombre de campo plano."""
    
    entity, channel, flow, hour = col_multi_index
    field_title = "tc_ars_{coin}_{entity}_{channel}_{flow}_{hour}".format(
        coin=coin,
        entity=entity,
        channel=channel,
        flow=flow,
        hour=hour
    )
    return field_title
```

Creo el CSV de series para dolar.

```python
df_pivot_dolar = df_panel[df_panel.coin == "dolar"].pivot_table(
    index="indice_tiempo", 
    columns=["entity", "channel", "flow", "hour"], 
    values="value"
)
df_pivot_dolar.head()
```

```python
flatten_columns = [create_field_title(col, "dolar") for col in df_pivot_dolar.columns]
df_pivot_dolar.columns = flatten_columns
df_pivot_dolar.reset_index(inplace=True)
df_pivot_dolar.head()
```

```python
df_pivot_dolar.to_csv("tipos-cambio-dolar-entidades-financieras-series.csv", encoding="utf8", index=False)
```
